```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 1. Load Boston dataset from the scikit-learn library

```python
df = pd.read_csv("input/Boston.csv")
df.head()
```

preprocess

```python
df = df.drop('Unnamed: 0',axis=1)
df.rename(columns={'medv':'price'},inplace=True)
```

```python
df.corr()
```

## 2. This dataset has 13 features for each house. According to data correlation and your analysis, choose 2 features for X, for example: 
##    X = area and age
##    and choose 1 target for Y:
##    Y = price

```python
x_train=np.array((df['rm'],df['zn']))
y_train=np.array(df['price'])
x_train=x_train.T
```

## 3. Split the dataset to train and test with train_test_split function.

```python
from sklearn.model_selection import train_test_split

X_train, X_test,Y_train, Y_test=train_test_split(x_train,y_train,test_size=0.2)
```

## 4. Solve the regression problem on 2 dimensions (2 features) using your Object Oriented LLS.

```python
from lls_class import LLS

lls=LLS()
w=lls.fit(X_train,Y_train)
w[0]
```

## 5. Show results on a 3D plot and draw the y_pred as a plane on it

```python

X1 = X_train[:,0]
X2 = X_train[:,1]

x1, x2 = np.meshgrid(X1, X2)
eq = w[0] * x1 + w[1] * x2 

fig = plt.figure()

ax = fig.add_subplot(111,projection='3d')
ax.scatter(X1 , X2 , Y_train,  color='green',alpha=0.5)
ax.plot_surface(x1, x2, eq)
ax.set_xlabel('rm')
ax.set_ylabel('zn')
ax.set_zlabel('price')
plt.show()
```

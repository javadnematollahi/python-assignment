```python
from myknn import KNN
accs=[3,5,7,9,11]
accuracy=[]
for i in accs:
    myknn1=KNN(i)
    myknn1.fit(np.array(x_train['shouldercircumference']),np.array(y_train))
    accuracy.append(myknn1.evaluate(np.array(x_test['shouldercircumference']),np.array(y_test)))
```

```python
Y_pred=myknn1.predict(np.array(x_test['shouldercircumference']))
confusion_matrix= np.zeros((2,2))
```

```python
y=y_test.reset_index()
```

```python

for i in range(len(Y_pred)):
    if y['Gender'][i]==0 and Y_pred[i]==0:
        confusion_matrix[0][0]+=1
    elif y['Gender'][i]==0 and Y_pred[i]==1:
        confusion_matrix[0][1]+=1
    elif y['Gender'][i]==1 and Y_pred[i]==0:
        confusion_matrix[1][0]+=1
    elif y['Gender'][i]==1 and Y_pred[i]==1:
        confusion_matrix[1][1]+=1
```

```python
from sklearn import metrics

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = ['women', 'men'])
cm_display.plot()
plt.title("confusion matix for myknn")
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier


XX_train=[]
x=x_train['shouldercircumference'].reset_index()
for i in range(len(x_train)):
    XX_train.append([x['shouldercircumference'][i]])

print(XX_train)
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(XX_train,(y_train))

XX_test=[]
x=x_test['shouldercircumference'].reset_index()
for i in range(len(x_test)):
    XX_test.append([x['shouldercircumference'][i]])

YY_pred=neigh.predict(XX_test)
```

```python
from sklearn.metrics import accuracy_score
accs=[3,5,7,9,11]
acc=[]
for i in accs:
    neigh = KNeighborsClassifier(n_neighbors=i)
    neigh.fit(XX_train,y_train)
    YY_pred=neigh.predict(XX_test)
    acc.append(accuracy_score(y_test, YY_pred))
print(acc)
```

```python
plt.plot(accs,accuracy,marker='o')
plt.plot(accs,acc,marker='o')
plt.xlabel("K")
plt.ylabel("ACCURACY")
plt.legend(["myknn","Sklearn"])
plt.title("accuracy for different K two knn classifier")
plt.show()
```

```python
from sklearn import metrics

confusion_matrix = metrics.confusion_matrix(y_test, YY_pred)

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = ["women", "men"])

cm_display.plot()
plt.title("confusion matix for Sklearn")
plt.show()
```

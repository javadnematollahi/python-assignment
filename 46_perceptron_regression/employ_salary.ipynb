## A. Make a linear dataset using scikit-learn library

```python
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt 

x, y, coef = datasets.make_regression(n_samples=100,#number of samples
                                      n_features=1,#number of features
                                      n_informative=1,#number of useful features 
                                      noise=10,#bias and standard deviation of the guassian noise
                                      coef=True,#true coefficient used to generated the data
                                      random_state=0) #set for same data points for each run

# Scale feature x (years of experience) to range 0..20
x = np.interp(x, (x.min(), x.max()), (0, 20))

# Scale target y (salary) to range 20000..150000 
y = np.interp(y, (y.min(), y.max()), (20000, 150000))
print(coef)
plt.ion() #interactive plot on
plt.plot(x,y,'.',label='training data')
plt.xlabel('Years of experience');plt.ylabel('Salary $')
plt.title('Experience Vs. Salary')
```

## B. Split data to train and test datasets

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test= train_test_split(x,y,shuffle=True ,test_size=0.2)
```

## C. Implement the perceptron algorithm as a class

```python
from perceptron import Perceptron

per=Perceptron(0.0001,0.1,4)
```

## D. Fit your model on the employee's salary dataset

```python
_,_,loss_train=per.fit(x_train,y_train)

Y_pred=per.predict(x_train)
```

## E. Plot data graph and loss graph as 2 subplots in 1 window

```python

fig , (ax1,ax2) =plt.subplots(1,2)
fig.set_figwidth(12)
ax1.scatter(x_train,y_train,color='blue')
ax1.plot(x_train,Y_pred,color='red')
ax1.set_xlabel("Years of experience")
ax1.set_ylabel("Salary $")
ax1.set_title('Experience Vs. Salary(Epoch=4)')
# losses=per.evaluate(x_test,y_test)
# print(losses)
ax2.plot(loss_train,color='green')
ax2.set_xlabel("epochs")
ax2.set_ylabel("Loss")
ax2.set_title('Loss vs epochs for train(Epoch=4)')
```

## F. Change hyperparameters for get best result

```python
per=Perceptron(0.0001,0.1,25)
_,_,loss_train=per.fit(x_train,y_train)

Y_pred=per.predict(x_train)

fig , (ax1,ax2) =plt.subplots(1,2)
fig.set_figwidth(12)
ax1.scatter(x_train,y_train,color='blue')
ax1.plot(x_train,Y_pred,color='red')
ax1.set_xlabel("Years of experience")
ax1.set_ylabel("Salary $")
ax1.set_title('Experience Vs. Salary(Epoch=25)')
# losses=per.evaluate(x_test,y_test)
# print(losses)
ax2.plot(loss_train,color='green')
ax2.set_xlabel("epochs")
ax2.set_ylabel("Loss")
ax2.set_title('Loss vs epochs for train(Epoch=25)')
```

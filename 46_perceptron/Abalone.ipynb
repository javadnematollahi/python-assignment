## A. Fit your perceptron algorithm on the abalone dataset

```python
import numpy as np
import pandas as pd
from sklearn import datasets
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from perceptron import Perceptron

df= pd.read_csv('input/abalone.csv')
df['Sex'].replace(["F","M","I"],[0,1,2],inplace=True)
df.head()
```

```python
df.corr()
```

```python
df=df.drop(['Sex'],axis=1)

y=df['Length'].values
x=df['Diameter'].values
```

```python
X_train, X_test,Y_train, Y_test=train_test_split(x,y,test_size=0.2)
X_train=X_train.reshape(-1,1)
X_test=X_test.reshape(-1,1)
Y_train=Y_train.reshape(-1,1)
Y_test=Y_test.reshape(-1,1)
```

```python
per=Perceptron(0.001,0.1,4)
_,_,loss=per.fit(X_train,Y_train)
Y_pred=per.predict(X_train)
```

## B. Plot data graph and loss graph as 2 subplots in 1 window

```python
fig , (ax1,ax2) =plt.subplots(1,2)
fig.set_figwidth(12)
ax1.scatter(X_train,Y_train,color='blue')
ax1.plot(X_train,Y_pred,color='red')
ax1.set_xlabel("Diameter")
ax1.set_ylabel("Length")
ax1.set_title('Length Vs. Diameter(Epoch=4)')
# losses=per.evaluate(x_test,y_test)
# print(losses)
ax2.plot(loss,color='green')
ax2.set_xlabel("epochs")
ax2.set_ylabel("Loss")
ax2.set_title('MAE vs epochs for train\n(Epoch=4, learning rate w=0.001, learning rate b=0.1)')
```
